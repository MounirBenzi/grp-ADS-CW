#### Linear Kernel

```python
for dep in Y.columns:
    screentime_var = X.values.tolist()
    depression_var = Y[dep].values.tolist()
    X_train, X_test, Y_train, Y_test = train_test_split(screentime_var, depression_var, test_size=0.5, random_state=0)
    
    smote_strat={}
    under_strat={}
    
    for category in Counter(Y_train).keys():
        if Counter(Y_train)[category] > round(len(Y_train)/(len(Counter(Y_train)))):
            smote_strat[category] = Counter(Y_train)[category]
            under_strat[category] = (round(len(Y_train)/(len(Counter(Y_train)))))-1
        else:
            smote_strat[category] = (round(len(Y_train)/(len(Counter(Y_train)))))-1
            under_strat[category] = (round(len(Y_train)/(len(Counter(Y_train)))))-1
            
    smote = SMOTE(sampling_strategy=smote_strat, k_neighbors=1)
    under = RandomUnderSampler(sampling_strategy=under_strat)

    steps = [('o', smote),('u',under)]
    pipeline = Pipeline(steps=steps)
    x_resample, y_resample = pipeline.fit_resample(X_train, Y_train)
    
    print("y_resample")
    print(Counter(y_resample))
    
    linear_svm = svm.SVC(kernel='linear', C=10).fit(x_resample, y_resample)
    print(linear_svm.score(X_test, Y_test))
```
