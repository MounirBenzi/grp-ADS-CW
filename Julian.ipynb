```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
breast = load_breast_cancer()
breast_data = breast.data
breast_labels = breast.target
breast_dataset = pd.DataFrame(breast_data)
features = breast.feature_names
print(features)
breast_dataset.head()
```

```python
x = breast_dataset.values
x = StandardScaler().fit_transform(x) # normalizing the features
x.shape
np.mean(x),np.std(x)
```

```python
normalised_breast = pd.DataFrame(x)
normalised_breast.tail()
```

```python
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
principal_breast_Df = pd.DataFrame(data = principalComponents_breast, columns = ['principal component 1', 'principal component 2'])
principal_breast_Df.tail()

pca_breast3d = PCA(n_components=3)
principalComponents_breast3d = pca_breast3d.fit_transform(x)
principal_breast_Df3d = pd.DataFrame(data = principalComponents_breast3d, columns = ['PC1', 'PC2','PC3'])
principal_breast_Df3d.tail()
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
print('Explained variation per principal component: {}'.format(pca_breast3d.explained_variance_ratio_))
```

```python
fig=plt.figure()
#plt.figure(figsize=(10,10))
#plt.xticks(fontsize=12)
#plt.yticks(fontsize=14)
#plt.xlabel('Principal Component - 1',fontsize=20)
#plt.ylabel('Principal Component - 2',fontsize=20)
#plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
#targets = ['Benign', 'Malignant']
#colors = ['r', 'g']
principal_breast_Df.plot(kind='scatter',x='principal component 1',y='principal component 2',color='red')
ax = fig.add_subplot(111, projection='3d')
data = principal_breast_Df3d.to_numpy()
print(data)
ax.scatter(data[:,0], data[:,1], data[:,2])
```
