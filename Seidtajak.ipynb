```python
%matplotlib widget
from ipywidgets import FloatSlider, IntSlider, interact, interact_manual

import pandas as pd
import numpy as np
import seaborn as sns
from IPython.display import display

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib

# read csv datafile into pandas table
df = pd.read_csv("maps-synthetic-data-v1.1.csv")
pd.options.display.max_columns = None
display(df)
#df.head()
```

```python
# Show first 10 columns, the missing and avaliable data
cols = df.columns[:] # first 10 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
# Finding the percentages of how missing to avaliable per col
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# first create missing indicator for features with missing data
for col in df.columns:
    missing = df[col].isnull()
    num_missing = np.sum(missing)
    #print(num_missing)
    
    # if data missing > 0, change the name of the co
    if num_missing > 0:  
        #print('created missing indicator for: {}'.format(col))
        df['{}_ismissing'.format(col)] = missing
```

```python
# then based on the indicator, plot the histogram of missing values
ismissing_cols = [col for col in df.columns if 'ismissing' in col]
df['num_missing'] = df[ismissing_cols].sum(axis=1)

df['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(figsize = (13,5), x = 'index', y = 'num_missing')
```

```python
# drop colsthi with a lot of missing values.
def dropCol(thresholds):
    count = 0
    drop_list = []
    for col in df.columns:
        pct_missing = np.mean(df[col].isnull())
        pct_missing = round(pct_missing*100)
        #print(pct_missing)
              
        if pct_missing > thresholds:
            count = count + 1
            df.drop(col, axis=1)
            drop_list.append(col)
            
    print('Total column drop is: ' + str(count))
    print('Columns which has too many N/A: \n', drop_list)
interact_manual(dropCol, thresholds=IntSlider(min=0, max=100));
```
